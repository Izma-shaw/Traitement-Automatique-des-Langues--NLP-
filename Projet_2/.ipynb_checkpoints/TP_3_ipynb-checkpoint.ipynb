[Q4] Ecrire la fonction qui calcule les K plus proches voisins d'un mot de référence sans recourir à la méthode wv.most_similar(). Vous devrez calculer le cosinus du mot de référence avec tous les mots du modèle qui se trouvent dans wv.vectors.

```python
# insérer votre code ici

def k_nearest_neighbors(word_reference, word_vectors, k=5):
    if word_reference not in word_vectors.key_to_index:
        raise ValueError(f"Le mot '{word_reference}' n'est pas présent dans le modèle.")

    # Récupérer l'index du mot de référence
    reference_index = word_vectors.key_to_index[word_reference]

    # Récupérer le vecteur du mot de référence
    reference_vector = word_vectors.vectors[reference_index]

    # Calculer les similarités cosinus avec tous les autres mots
    cosine_similarities = np.dot(word_vectors.vectors, reference_vector) / (np.linalg.norm(word_vectors.vectors, axis=1) * np.linalg.norm(reference_vector))

    # Obtenir les indices des K plus grandes similarités
    k_neighbors_indices = np.argsort(cosine_similarities)[::-1][:k]

    # Récupérer les mots et les valeurs de similarité correspondantes
    k_neighbors_words = [word_vectors.index_to_key[i] for i in k_neighbors_indices]
    k_neighbors_similarities = [cosine_similarities[i] for i in k_neighbors_indices]

    return k_neighbors_words, k_neighbors_similarities

# Exemple d'utilisation
reference_word = "cup"
k_neighbors, k_similarities = k_nearest_neighbors(reference_word, wv)


# Imprimer les mots et les valeurs de similarité dans le format demandé
for word, similarity in zip(k_neighbors, k_similarities):
    print(f"{word} [{similarity}]")
```

```python
# Fonction pour effectuer une analogie et formatter les résultats
def perform_analogy(positive, negative, topn=10):
    result = wv.most_similar(positive=positive, negative=negative, topn=topn)
    return result

# Tester plusieurs analogies
capital_of_bulgaria = perform_analogy(['bulgaria', 'capital'], [])
luxury_brand_of_nissan = perform_analogy(['nissan', 'luxury'], [])
past_tense_of_carry = perform_analogy(['carry', 'past'], [])

# Formatter et afficher les résultats
def format_and_print_result(analogy_name, result):
    formatted_result = [(word, similarity) for word, similarity in result]
    print(f"{analogy_name}: {formatted_result}")

format_and_print_result("Capitale de la Bulgarie", capital_of_bulgaria)
format_and_print_result("Marque de luxe de Nissan", luxury_brand_of_nissan)
format_and_print_result("Prétérite du verbe 'carry'", past_tense_of_carry)
```

3 - Visualisation des plongements lexicaux
------

Il existe plusieurs méthodes de visualisation des données en N dimensions. Elles permettent de réduire la dimensionnalité de l'espace pour la rendre visible dans une espace à 2 ou à 3 dimensions. L'une de ces méthodes est la [t-SNE](https://distill.pub/2016/misread-tsne/).

```python
import matplotlib.pyplot as plt
import random

def reduce_dimensions(wv, n):
    num_dimensions = 2  # final num dimensions (2D, 3D, etc)

    # extract the words & their vectors, as numpy arrays
    vectors = np.asarray(wv.vectors[:n])
    labels = np.asarray(wv.index_to_key[:n])

    # reduce using t-SNE
    tsne = TSNE(n_components=num_dimensions, learning_rate='auto', init='random')
    vectors = tsne.fit_transform(vectors)

    x_vals = [v[0] for v in vectors]
    y_vals = [v[1] for v in vectors]
    return x_vals, y_vals, labels

def plot_with_matplotlib(x_vals, y_vals, labels):
    plt.figure(figsize=(12, 12))
    plt.scatter(x_vals, y_vals)

    # Label randomly subsampled 100 data points
    random.seed(0)
    indices = list(range(len(labels)))
    selected_indices = random.sample(indices, 100)
    for i in selected_indices:
        plt.annotate(labels[i], (x_vals[i], y_vals[i]))

x_vals, y_vals, labels = reduce_dimensions(wv, 1000)
plot_with_matplotlib(x_vals, y_vals, labels)
```

4 - PROJET : Réécrire un texte à partir de tranformations basées sur les plongements lexicaux
------

[Projet Q1] Réécrire un texte donné en anglais de sorte à ce que chaque substantif du texte soit remplacé par le Kème mot le plus proche dans le plongement lexical. Vous générerez le nuage de mots correspondant à partir du texte réécrit.

```python
# insérer votre code ici
#!python -m spacy download en_core_web_lg



# Chargez le modèle de traitement grammatical SpaCy
nlp = spacy.load('en_core_web_lg')

def remplacer_substantifs(texte, k):
    # Analysez le texte avec SpaCy
    doc = nlp(texte)

    # Initialisez une liste pour stocker le texte modifié
    texte_modifie = []
    voisin_list= []

    # Parcourez les tokens dans le texte
    for token in doc:
        # Vérifiez si le token est un substantif
        if token.pos_ == 'NOUN'and token.text in wv.key_to_index:
            # Obtenez les k mots les plus similaires du modèle Word2Vec
            voisins = wv.most_similar(positive=[token.text], topn=k)

            # Ajoutez le mot le plus similaire au texte modifié
            texte_modifie.append(voisins[0][0])
            voisin_list.append(voisins[0][0])
        else:
            # Ajoutez les autres types de mots tels quels
            texte_modifie.append(token.text)

    # Rejoignez les mots modifiés pour former le nouveau texte
    texte_final = ' '.join(texte_modifie)

    return texte_final,voisin_list

# Exemple d'utilisation
text_path1 = "gender-en.txt"

text1 = ""
with open(text_path1, encoding='utf-8') as f:
    text1 = f.read()

k_voisins = 3
texte_modifie = remplacer_substantifs(text1, k_voisins)[1]

print("Texte original :")
print(text1)
print("\nTexte modifié :")
print(texte_modifie)
```

```python
# insérer votre code ici


def generate_wordcloud(freq_list):
    font_path= "/usr/share/fonts/truetype/fonts-japanese-gothic.ttf"
    wordcloud = WordCloud(max_font_size=100,font_path=font_path).generate_from_text(freq_list)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    

texte = ' '.join(texte_modifie)

generate_wordcloud(texte)

```

5 - PROJET : Redéfinir l'algorithme Lesk
------

[Projet Q3] L'algorithme de Lesk (https://en.wikipedia.org/wiki/Lesk_algorithm) permet de désambiguiser le sens d'un mot dans un contexte défini. Ainsi il retourne le sens correspondant à un mot dans un contexte donné. La fonction OVERLAP implémentée dans l'algorithme Lesk permet de choisir le meilleur sens en fonction du cardinal de l'intersection des mots entre la définition du sens et le contexte textuel.

Ici, vous proposerez de faire cette sélection à partir d'une représentation sémantique continue des phrases. Chaque phrase devra être donc représentée par un vecteur sémantique. Vous pourrez utiliser le vecteur moyen de tous les mots dans une phrase (average pooling) ou des techniques plus abouties comme celle proposé par [(Hovy, 2025)](https://aclanthology.org/P15-1073.pdf). Le sens d'un mot dans son contexte sera donc celui qui maximise la similarité entre la représentation sémantique de la définition et celle du contexte.
